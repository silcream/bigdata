```python
result1 = []

for page in range(1, 5):
    url = 'http://www.edavinci.co.kr/pg/bbs/board.php?bo_table=store&page='
    soup = BeautifulSoup(html, 'html.parser')
    resp = requests.get(url)
    html = resp.content
    table = soup.find('table')
    td = table.find('table')
    tr_tags = td.find_all('tr')

    for tr in tr_tags :
        if len(tr.find_all('td')) > 0:
            td_tags = tr.find_all('td')

            sido = td_tags[1].string
            name = td_tags[2].string
            phone = td_tags[3].string
            addr = td_tags[4].string.strip()
            result1.append([sido, name, addr, phone])
    print(f'-----------{page} page completed-----')
```
