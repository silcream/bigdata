# 파일오픈

```python
f = open('test.txt', 'w')
f.write('contents')
f.close()
```

readline : 그냥 반환

readlines : list로 반환

```python
f2 = open('test.txt', 'r')
line = f2.readlines()
print(line)
f2.close()
```

```python
f3 = open('test.txt', 'r')
lines = f3.readlines()
for line in lines:
    print(line)
f3.close()
```

자동으로 열리고 닫힘

```python
with open('test.txt', 'r') as f4 :
    print(f4.readlines())    
```
