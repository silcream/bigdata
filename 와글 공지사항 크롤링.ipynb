```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
url = 'https://www.changwon.ac.kr/portal/main.do'
driver.get(url)
result = []

time.sleep(1) #쉬기
driver.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/div/div/div[1]/a').click()
time.sleep(1) #쉬기
for page in range(1, 10):
    try : 
        driver.execute_script("goPaging(%d)" % page) #클릭

        time.sleep(1) #쉬기
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        tr = soup.select('tr')
        for i in range(5, 15):
            number = tr[i].select('td')[0].string
            title = tr[i].select('td')[1].a.string.strip()
            date = tr[i].select('td')[2].string
            see = tr[i].select('td')[3].string
            
            result.append([number, title, date, see])
        
        
    except Exception as e:
        time.sleep(1)
        driver.get(url)
        print(e)
```
