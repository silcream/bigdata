```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
url = 'https://www.coffeebeankorea.com/store/store.asp'
driver.get(url)
result = []


for i in range(1, 100):
    try:
        driver.execute_script("storePop2(%d)" % i) #클릭
        time.sleep(1) #쉬기
        html = driver.page_source
        soupCB1 = BeautifulSoup(html, 'html.parser')
        title = soupCB1.select('div.store_txt > h2')[0].string
        addr = soupCB1.select('div.store_txt > table.store_table > tbody > tr > td')[2].text
        phone = soupCB1.select('div.store_txt > table.store_table > tbody > tr > td')[3].string

        print(title, addr, phone)

        driver.find_element(By.ID, 'closeBtn1ActiveBtn').send_keys(Keys.ENTER)
        result.append([title, addr, phone])

    except Exception as e:
        time.sleep(1)
        driver.get(url)
        print(e)
```
