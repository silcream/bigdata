```python
url1 = 'http://www.gnnews24.kr/news/articleList.html?page=%d&total=13203&sc_section_code=&sc_sub_section_code=&sc_serial_code=&sc_area=&sc_level=&sc_article_type=&sc_view_level=&sc_sdate=&sc_edate=&sc_serial_number=&sc_word=&box_idxno=&sc_multi_code=&sc_is_image=&sc_is_movie=&sc_order_by=E&view_type=sm' 
resp1 = requests.get(url1)
html1 = resp1.content
soup1 = BeautifulSoup(html1, 'html.parser')
div_tag = soup1.find('div', {'id' : 'user-gnb'})
ul_tag1 = div_tag.find('ul', {'id' : 'userLogin'})
lili = ul_tag1.find_all('li')[3].string
year = lili[0:4]
```

```python
results = []
for page in range(1, 5):
    url = 'http://www.gnnews24.kr/news/articleList.html?page=%d&total=13203&sc_section_code=&sc_sub_section_code=&sc_serial_code=&sc_area=&sc_level=&sc_article_type=&sc_view_level=&sc_sdate=&sc_edate=&sc_serial_number=&sc_word=&box_idxno=&sc_multi_code=&sc_is_image=&sc_is_movie=&sc_order_by=E&view_type=sm' %page 
    resp = requests.get(url)
    html = resp.content
    soup = BeautifulSoup(html, 'html.parser')
    
    section_tag = soup.find('div', {'id' : 'sections'})
    ul_tag = section_tag.find('section', {'id' : 'section-list'}).find('ul')
    li_tag = ul_tag.find_all('li')
    for li in li_tag:
        title = li.find('h4').find('a').string
        abstract = li.p.a.string.strip()
        rep_name = li.find_all('em')[1].string
        time = li.find_all('em')[2].string
        link = li.find('h4').find('a')['href']
        year = year

        results.append([title, abstract, rep_name, time, link, year])
```

```python
results
```

```python
result_df = pd.DataFrame(results, columns=['title', 'abstract', 'rep_name', 'time', 'link', 'year'])
result_df
```
